7. Tampilkan `ID`, `AGE_RANGE`, `INCOME`, dan `EDUCATION` pemegang polis dengan `SPEEDING_VIOLATIONS` tertinggi.

8. Filter `df` untuk `VEHICLE_YEAR` sebelum tahun 2015, simpan dalam variabel `df_filtered` dan tampilkan hasilnya.

```python
df_filtered = df.loc[df["VEHICLE_YEAR"] == "before 2015"]
df_filtered
```

9. Tampilkan rata-rata kolom `OUTCOME` untuk tiap `AGE_RANGE` dari `df_filtered` yang didefinisikan pada nomor sebelumnya.

```python
df_filtered.groupby("AGE_RANGE")["OUTCOME"].mean()
```

10. Tampilkan rata-rata kolom `OUTCOME` untuk tiap pasangan `AGE_RANGE` dan `INCOME` dari `df_filtered`. **Tampilkan sebagai `DataFrame`**.

```python
df_filtered.groupby(["AGE_RANGE", "INCOME"])["OUTCOME"].mean().reset_index()
```
