# **Exercise: Pandas Merging**

Gunakan datasets:
- `audi.csv`
- `hyundi.csv`
- `toyota.csv`

Load ketiga dataset di atas, kemudian simpan masing-masing dataset ke dalam sebuah variable!

```python
import numpy as np
import pandas as pd
```

Dataset: https://www.kaggle.com/adityadesai13/used-car-dataset-ford-and-mercedes?select=toyota.csv

```python
df_audi = pd.read_csv(r'D:\JCDS\Dataset\used car data set\audi.csv', index_col=0)
df_audi.head()
```

```python
df_hyundai = pd.read_csv(r'D:\JCDS\Dataset\used car data set\hyundi.csv', index_col=0)
df_hyundai.head()
```

```python
df_toyota = pd.read_csv(r'D:\JCDS\Dataset\used car data set\toyota.csv', index_col=0)
df_toyota.head()
```

## **No.1**

- Jika Anda perhatikan, nama kolom 'tax(£)' di dataset hyundai berbeda dengan di dataset lainnya.
- Ubah nama kolom tersebut menjadi 'tax' agar sama dengan di dataset yang lain.

```python
df_hyundai = df_hyundai.rename(columns={"tax(£)" : "tax"})
df_hyundai.head()
```

## **No.2**

- Buat kolom baru bernama 'brand' dengan isi nama brand mobilnya, letakkan pada kolom paling kiri.
- Misalnya untuk data frame audi, isi kolom ‘brand’ dengan ‘Audi’!

```python
df_audi.insert(0, "brand", "Audi")
df_audi.head()
```

```python
df_toyota.insert(0, "brand", "Toyota")
df_toyota.head()
```

```python
df_hyundai.insert(0, "brand", "Hyundai")
df_hyundai.head()
```

## **No.3**

- Gabungkan ketiga data frame tersebut menjadi sebuah data frame bernama df_car. Atur agar indexnya tidak ada yang berulang.


```python
df_car = pd.concat([df_audi, df_hyundai, df_toyota], axis=0).reset_index()
df_car
```

## **No.4**

- Pada kolom 'model' di sebagian data, ada whitespace tambahan di awal atau di akhir yang tidak berguna.
- Hapus whitespace tersebut dari semua data pada kolom 'model'!

```python
df_car["model"] = df_car["model"].str.strip()
```

```python
df_car["model"].unique()
```

## **No.5**

- Dari df_car, tampilkan rata-rata pajak/tax berdasarkan brand dan transmission. Lalu urutkan dari tax yang paling tinggi.


```python
df_car.groupby(["brand", "transmission"])["tax"].mean().reset_index().sort_values("tax", ascending=False)
```

## **No.6**

- Anda diminta untuk mencari mobil Toyota Camry yang harganya (price) di bawah rata-rata harga Camry dan jarak tempuhnya (mileage) di bawah rata-rata mileage Camry juga.
- Tampilkan mobil yang sesuai dengan kriteria tersebut!

```python
camry_price_avg = df_toyota[df_toyota["model"].str.strip().str.lower() == "camry"]["price"].mean()
camry_price_avg
```

```python
camry_price_avg = df_toyota[df_toyota["model"].str.strip().str.lower() == "camry"]["price"].mean()
camry_mileage_avg = df_toyota[df_toyota["model"].str.strip().str.lower() == "camry"]["mileage"].mean()

df_toyota[(df_toyota["model"].str.strip() == "Camry") &
          (df_toyota["price"] < camry_price_avg) &
          (df_toyota["mileage"] < camry_mileage_avg)]
```

## **No.13**

Buat dataframe baru dengan pivot table berisi rata-rata tax tiap transmission dan fuelType

## **No.15**
- **Filter data yang berasal dari 3 kategori dengan jumlah film terbanyak**

- **Dari data yang telah difilter, tampilkan rating film beserta rata-rata durasi rentalnya**

- **Tampilkan dengan nama kolom berikut: `rating`, `avg_durasi_rental`**
- **Urutkan dari `avg_durasi_rental` terbesar**


```python
film_path = "D:\\JCDS\\Dataset\\film.csv"

film = pd.read_csv(film_path, index_col=0)
film.head()
```

```python
category_path = "D:\\JCDS\\Dataset\\category.csv"

category = pd.read_csv(category_path)
category.head()
```

```python
film_category_path = "D:\\JCDS\\Dataset\\film_category.csv"

film_category = pd.read_csv(film_category_path)
film_category.head()
```

```python
merge1 = pd.merge(category, film_category, on="category_id")
merge2 = pd.merge(film, merge1, on="film_id")

merge2
```

```python
merge2.groupby("rating")["rental_duration"].mean().reset_index().sort_values("rental_duration", ascending=False)
```
