## **No. 1b**

* Buatlah sebuah array 1 dimensi dari data random integer sebanyak “n”, dimana “n” tersebut adalah sebuah input !
* Batasan random juga ditentukan dalam input.
* Ubahlah array 1 dimensi tersebut ke dalam bentuk 2 dimensi atau matrix!
* Setelah diubah ke bentuk 2 dimensi, setiap barisnya baliklah urutan datanya! Misalnya dari 1, 2, 3 menjadi 3, 2, 1.


```python
#Jawab
while True:
    # Step 1. meminta input Banyak data, batas
    input_n = input("Silahkan masukan banyak data bilangan random: ")
    input_stop_val = input("Silahkan masukan angka selesai random interger: ")
    input_start_val = input("Silahkan masukan angka mulai random interger: ")
    
    # Step 2. validasi format hanya digit
    if input_n.isdigit() and input_stop_val.isdigit() and input_start_val.isdigit():
        input_n = int(input_n)
        input_stop_val = int(input_stop_val)
        input_start_val = int(input_start_val)
        
        # Step 3. validasi agar value banyak data dan start tidak melebihi stop
        if input_n > 0 and input_start_val < input_stop_val:
            array_1D = np.random.randint(input_start_val, input_stop_val, input_n) # (Start, Stop, Step)
            break
    else:
        print("Input tidak valid. coba lagi!")
    

array_1D # Banyak data: 40, Mulai: 1, Selesai: 21
```

```python
# reshape to 2D
array_2D = array_1D.reshape(8, 5)
array_2D
```

```python
# reverse order by indexing
array_2D_reverse = array_2D[:, ::-1]
array_2D_reverse

# yang di reverse hanya kolom jadi step -1 hanya ada di bagian kolom
```

## **No. 3**
Sekarang, dari data mentah di atas, coba buat dataset seperti di bawah ini.  
Perhatikan bahwa:  
* kalian bisa mendapatkan kolom `Gender` dari `Title` yang berasal dari kolom `Name`. Tergantung dari cara kalian membuat kolom gender, jumlah Female/Male yang kalian dapat bisa berbeda 1 dari contoh. 
* kolom `Alone` berisi alone/not alone sesuai kondisi penumpang, berdasarkan kolom yang ada.

```python
df.head()
```

```python
# 1. ambil title dari name menggunakan regex
df["Title"] = df["Name"].str.extract(r",\s*(\w+)\.")
df.head()
```

```python
# First Name = setelah spasi dari titik title hingga akhir
df["First Name"] = df["Name"].str.extract(r'\.\s*(.+)$')
df.head()

# . -> placeholder, ambil karakter apapun
# \s* nol atau lebih spasi
# (.+) -> ambil semua karakter semuanya
# $ -> sampai akhir string
```

```python
# Last name = dari awal hingga sebelum koma 
df["Last Name"] = df["Name"].str.extract(r'^([^,]+)')
df.head()

# ^ -> ambil start dari awal string
# ([^,]) -> karakter selain koma ( , )
# + -> ambil satu atau lebih karakter
# ^([^,]+) -> Ambil satu atau lebih karakter dari awal string hingga sebelum koma
```

```python
# 2. GENDER dari TITLE
# Step 1. fungsi untuk menentukan gender dari title
def gender_title(title):
    if title in ["Mr", "Master", "Dr", "Rev", "Col", "Major", "Sir", "Capt", "Don", "Jonkheer"]:
        return "Male"
    else:
        return "Female"
        
```

```python
# Step 2. apply fungsi disetiap baris df
df["Gender"] = df["Title"].apply(gender_title)
df.head(2)
```

```python
# ALTERNATIF
df["Gender"] = df["Title"].apply(lambda title : "Make" if title in ["Mr", "Master", "Dr", "Rev", "Col", "Major", "Sir", "Capt", "Don", "Jonkheer"] else "Female")
df.head(2)
```

```python
# 3. Menentukan Alone dari SibSp dan Parch
# Step 1. fungsi cek
def alone_or_not(status):
    if status["SibSp"] == 0 and status["Parch"] == 0:
        return "Alone"
    else:
        return "Not Alone"
```

```python
# Step 2. apply fungsi disetiap baris
df["Alone"] = df.apply(alone_or_not, axis = 1) # axis = 1 --> agar bekerja baris per baris, 0 --> kolom per kolom
df.head()
```

```python
# ALTERNATIF
df["Alone"] = df.apply(
    lambda x: "Alone" if x["SibSp"] == 0 and x["Parch"] == 0 else "Not Alone", axis=1)
df.head()
```

```python
# 4. reorder kolom
df = df[["PassengerId", "Title", "First Name", "Last Name", "Gender", "Age", "Alone", "Pclass", "Fare", "Embarked", "Survived"]]
df
```

```python
# 1. Gender Vs Survivability
df.groupby("Gender")["Survived"].mean()
```

```python
# 2. Age Vs Survivability
def group_age(age):
    if age < 13:
        return "Child"
    elif age < 18:
        return "Teenager"
    elif age < 36:
        return "Young Adult"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"
```

```python
df["Age Group"] = df["Age"].apply(group_age)
df.groupby("Age Group")["Survived"].mean()
```

```python
# 3. Pclass Vs Survivability
df.groupby("Pclass")["Survived"].mean()
```

```python
# 4. Fare Vs Survivability
df["Fare"].describe()
```

```python
def fare_to_quartil(fare):
    if fare < 7.910400:
        return "ver low"
    elif fare < 14.454200:
        return "low"
    elif fare < 31.000000:
        return "high"
    elif fare < 512.329200:
        return "very high"
```

```python
df["Fare Level"] = df["Fare"].apply(fare_to_quartil)
```

```python
df.groupby("Fare Level")["Survived"].mean()
```
