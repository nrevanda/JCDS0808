## Task 1 - Stock price fluctuations

- Create subplots using `matplotlib.pyplot.subplots` for each company's stock price.
- Utilize matplotlib.pyplot.plot to depict stock prices as line plots.
- Add annotations with matplotlib.pyplot.annotate to highlight key events (e.g., earnings, product releases) that influenced stock prices.

```python
# Dataset
# ====================

# Random seed
np.random.seed(101)

# Simulated data for stock prices
time = list(range(1, 21))
stock_prices_company1 = np.random.normal(100, 5, 20)
stock_prices_company2 = np.random.normal(120, 8, 20)

# Creating a DataFrame
data = {
    'Time': time,
    'Company 1 Stock Price': stock_prices_company1,
    'Company 2 Stock Price': stock_prices_company2
}

df = pd.DataFrame(data)
```

```python
df
```

```python
fig, axes = plt.subplots(2, 1, figsize=(11.7, 10))

sns.lineplot(data=df,
             x='Time',
             y='Company 1 Stock Price',
             ax=axes[0],
             color='blue',
             label = 'Company 1')
axes[0].grid(True)
axes[0].set_title("Stock Prices for Company 1")
axes[0].legend(loc='lower left')


sns.lineplot(data=df,
             x='Time',
             y='Company 2 Stock Price',
             ax=axes[1],
             color='green',
             label = 'Company 2')
axes[1].grid(True)
axes[1].set_title("Stock Prices for Company 2")
axes[1].legend(loc='upper left')


plt.subplots_adjust(hspace=0.2)
plt.show()
```

```python
# Visualization
# ====================
```

Insight:
- kedua saham mengamalami tren naik lalu kembali naik pada rentang waktu 20.0 unit. keduanya mengamalami rebound pada waktu ke 15.0

## Task 2 - Relationship between study hours and exam scores

Buat visualisasi data dengan menggunakan data yang telah disediakan. Jalankan blok kode 'Dataset' untuk membangkitkan data.

```python
# Dataset
# ====================

# Random seed
np.random.seed(101)

num_students = 100

study_hours = np.random.normal(7, 2, num_students)
exam_scores = np.random.normal(75, 15, num_students)

# Adding noise to the data
study_hours += np.random.normal(0, 2, num_students)
exam_scores += np.random.normal(0, 5, num_students)

# Generating random grade levels and majors for students
grade_levels = np.random.choice(['Junior', 'Senior'], num_students)

# Create DataFrame
data = {
    'Study Hours': study_hours,
    'Exam Scores': exam_scores,
    'Grade Level': grade_levels,
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.figure(figsize=(11.7, 6.5))
sns.scatterplot(data=df, x='Study Hours', y='Exam Scores', hue='Grade Level')
sns.regplot(data=df, x='Study Hours', y='Exam Scores', scatter=False)

plt.title("Relitionship between Study Hours and Exam Scores")
plt.grid(True)
plt.legend(loc='upper left', title='Grade Level')
```

```python
# Visualization
# ====================
```

Insight:
- tidak ada tren korelasi antara study hours dengan exam scores
