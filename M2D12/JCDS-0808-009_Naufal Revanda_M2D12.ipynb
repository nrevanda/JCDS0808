### No. 3 
Nilai outlier adalah nilai yang berada di bawah Q1 - (1.5 * IQR) atau di atas Q3 + (1.5 * IQR).

a. Hitung dulu Q1, Q2, dan IQR pada kolom Price (harga rumah)

b. Hitung ada berapa nilai outlier pada kolom Price!


```python
# a

# menghitung quantile
q1_price = df["Price"].quantile(0.25)
q2_price = df["Price"].quantile(0.5)
q3_price = df["Price"].quantile(0.75)

print(f"Q1 : {q1_price}")
print(f"Q2 : {q2_price}")
print(f"Q3 : {q3_price}")
```

```python
# menghitung Inter Quantile Range
iqr_price = q3_price - q1_price

print(f"IQR : {iqr_price}")
```

```python
# menghitung boundary
lower_boundary = q1_price - (1.5*iqr_price)
upper_boundary  = q3_price + (1.5*iqr_price)

print(f"Upper boundary: {upper_boundary}")
print(f"Lower boundary: {lower_boundary}")
```

```python
# menghitung jumlah outlier

jumlah_outlier_bawah = df[df["Price"] < lower_boundary]
jumlah_outlier_atas = df[df["Price"] > upper_boundary]

jumlah_outlier = len(jumlah_outlier_atas) + len(jumlah_outlier_bawah)
jumlah_outlier
```

### No. 9 Uji hipotesis apakah metode pembelian/Method berkaitan dengan tipe rumah/ Type (dependen) atau tidak saling berkaitan (independen)!

Hypothesis:
- Ho : The two variables are independent
- Ha : The two variables are dependent.


```python
sns.histplot(data=df,
             x="Method",
            )
```

```python
sns.histplot(data=df,
             x="Type",
            )
```

```python
# Cek jumlah tiap kategori
method_type_count = pd.crosstab(df["Method"], df["Type"])
method_type_count
```

```python
from scipy.stats import chi2_contingency
```

```python
# Ho : The two variables are independent
# Ha : The two variables are dependent.

stat, pvalue, *_ = chi2_contingency(method_type_count)
pvalue

# Pvalue < 0.05
# tolak H0 --> variabel method dan type berkaitan (dependent)
```

```python
# secara visual
method_type_count_normalized = pd.crosstab(df["Method"], df["Type"], normalize=0)

sns.heatmap(method_type_count_normalized, 
            cmap="viridis",
            annot=True,
            vmin=0, vmax=1)

# heatmap menunjukan 
    # metode S dan SA paling sering dipake untuk tipe h (72%)
    # juga metode VB dan SP cenderung lebih tinggi dipakai untuk tipe u (~29%)
# artinya ada keterkaitan antara metode dengan tipe rumah --> mendukung HA
```

### No. 10 
Uji hipotesis apakah rata-rata harga rumah (Price) antar Regionname berbeda ataukah sama! 

Tentukan terlebih dahulu apakah persyaratan parametrik terpenuhi, dan gunakan metode uji hipotesis yang sesuai.

Hipotesis:
- Ho : μ1 = μ2 = … = μk (Rata-rata harga rumah antar Regionname sama)
- Ha : Setidaknya ada satu pasang Regionname yang rata-rata harga rumahnya tidak sama, tapi kita tidak tahu yang mana


```python
df.head()
```

```python
region_price_median = df.groupby("Regionname")[["Price"]].median()
region_price_median
```

```python
sns.barplot(data=df,
            x="Regionname",
            y="Price",
            estimator="median")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# uji distribusi normal
for region, group in df.groupby("Regionname"): # loop untuk setiap group berdasarkan regionname
    prices = group["Price"] #mengambi kolom price tiap group regionname
    if len(prices) >= 3: 
        stat, pvalue = shapiro(prices)
        print(f"{region}: p-value = {pvalue}")
        
# hanya Easter Metropolitan saja yang p-value > 0.05
# sebagian besar tidak terdistribusi normal --? bisa pakai uji non parametrik
```

```python
# karena data tidak normal, dan lebih dari 2 variabel. pakai Kruskal Wallis
from scipy.stats import kruskal
```

```python
region_prices = []

for region, gorup in df.groupby("Regionname"):
    prices = group["Price"]
    region_prices.append(prices)
```

```python
# Ho : μ1 = μ2 = … = μk (Rata-rata harga rumah antar Regionname sama)
# Ha : Setidaknya ada satu pasang Regionname yang rata-rata harga rumahnya tidak sama, tapi kita tidak tahu yang mana

stat, pvalue = kruskal(*region_prices)
pvalue

# pvalue > 0.05
# gagal menolak H0 --> tidak ada cukup bukti untuk menyatakan rata-rata harga rumah antara region berbeda
```

### No. 11
- A. dari group treatment, apakah spending pengguna setelah diberi promo lebih besar dari sebelum diberi promo? gunakan uji hipotesis yang sesuai
- B. dari group control, apakah spending pengguna setelah diberi promo lebih besar dari sebelum diberi promo? gunakan uji hipotesis yang sesuai

```python
df_aplikasi = pd.read_csv(r'D:\0. JCDS\Dataset\Sales_without_NaNs.csv')
df_aplikasi.head()
```

```python
# step 1. hitung selisih
df_aplikasi["Sales_diff"] = df_aplikasi["Sales_After"] - df_aplikasi["Sales_Before"]
```

```python
# step 2. cek distribusi normal secara visual
sns.histplot(data=df_aplikasi, x="Sales_diff")
```

```python
# step 2. cek distribusi normal secara statistik
stats, pvalue = shapiro(df_aplikasi["Sales_diff"])
pvalue

# pvalue < 0.05
# data tidak terdistribusi normal
```

```python
# step 3. filter group

treatment = df_aplikasi[df_aplikasi["Group"] == "Treatment"]
control = df_aplikasi[df_aplikasi["Group"] == "Control"]
```

```python
# step 4. uji hipotesis

# H0: sales setelah diberi promo = sebelum diberi promo
# HA: sales setelah diberi promo > sebelum diberi promo

# cocok menggunaka uji hipotesis wilcooxon test
# kondisi mirip 2 sample paired t-test -> membandingkan 1 kelompok dengan data berpasangan (Salse_Before Vs Sales_After)

from scipy.stats import wilcoxon

stats_treatment, pvalue_treatment = wilcoxon(treatment["Sales_After"],
                         treatment["Sales_Before"],
                         alternative = "greater")



stats_control, pvalue_control = wilcoxon(control["Sales_After"],
                         control["Sales_Before"],
                         alternative = "greater")

print(f"P-Value group treatment: {pvalue_treatment}")
print(f"P-Value group control: {pvalue_control}")

# Kedua group treatment maupun control memiliki nilai pvalue < 0.05
# berhasil menolak H0 --> promo berhasil meningkatkan penjualan, untuk group control, meskipun tidak diberi promo namun penjualan tetap meningkat
```
