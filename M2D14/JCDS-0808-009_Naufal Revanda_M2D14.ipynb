# **No.1**

Tampilkan visualisasi berupa heatmap yang menunjukkan nilai korelasi antara variabel 'Rooms', 'Price', 'Distance', 'BuildingArea', 'YearBuilt'! Variabel apa yang memiliki korelasi paling kuat dengan harga rumah (Price)?

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))

# Flatten axes agar mudah dipakai
axes = axes.flatten()

sns.histplot(data=df, x="Price", kde=True, ax=axes[0])
axes[0].set_title("Distribusi Harga (Price)")

sns.histplot(data=df, x="BuildingArea", kde=True, ax=axes[1])
axes[1].set_title("Distribusi Luas Bangunan (BuildingArea)")

sns.histplot(data=df, x="Distance", kde=True, ax=axes[2])
axes[2].set_title("Distribusi Jarak ke CBD (Distance)")

sns.histplot(data=df, x="YearBuilt", kde=True, ax=axes[3])
axes[3].set_title("Distribusi Tahun Dibangun (YearBuilt)")


plt.tight_layout()
plt.show()
```

```python
cols = ['Rooms', 'Price', 'Distance', 'BuildingArea', 'YearBuilt']
corr_df = df[cols].copy()
corr_df = corr_df.dropna()
corr_matrix = corr_df.corr(method='spearman')
corr_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f', square=True)
plt.title("Heatmap Korelasi Antar Variabel Properti")
plt.show()
```

# **No.2**

Tampilkan visualisasi yang menggambarkan perbandingan dari persebaran harga rumah (Price) berdasarkan tipe rumah (Type)! Rumah di harga berapa untuk sebuah rumah bertipe "t" (townhouse) bisa dikatakan sebagai outlier?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Type", y="Price", palette="Set2")
plt.title("Persebaran Harga Rumah Berdasarkan Tipe Rumah")
plt.xlabel("Tipe Rumah")
plt.ylabel("Harga Rumah")
plt.grid(True)
plt.show()
```

```python
harga_tipe_t = df[df['Type'] == 't']
harga_tipe_t['Price'].dropna()
harga_tipe_t
```

```python
q1 = harga_tipe_t['Price'].quantile(0.25)
q3 = harga_tipe_t['Price'].quantile(0.75)
iqr = q3 - q1

upper_bound = q3 + 2*iqr
lower_bound = q1 - 2*iqr

print(f'otulier bawah untuk harga rumah tipe t: {lower_bound}')
print(f'otulier atas untuk harga rumah tipe t: {upper_bound}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Baca data
store = pd.read_excel(r'C:\Users\Lenovo\Documents\Sample - Superstore.xlsx')

# Agregasi sales per kategori
category_sales = store.groupby("Category")["Sales"].sum()
labels = category_sales.index
sizes = category_sales.values
colors = plt.cm.Set2.colors  # Gunakan warna dari colormap Set3
color_dict = dict(zip(labels, colors))  # Pemetaan warna berdasarkan kategori
total_sales = category_sales.sum()

# Ambil top 5 product dan kategorinya
top_5_product = (
    store.groupby(['Product ID', 'Category'])['Sales']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
    .head(5)
)

# Buat kolom label baru dengan prefix berdasarkan kategori
prefix_map = {'Technology': 'TEC', 'Office Supplies': 'OFF', 'Furniture': 'FUR'}
top_5_product['Label'] = top_5_product.apply(
    lambda row: f"{prefix_map[row['Category']]}-{row['Product ID']}", axis=1
)

# Buat subplot
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

# --- Donut chart pada axes[0] ---
wedges, texts, autotexts = axes[0].pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=[color_dict[label] for label in labels],
    wedgeprops=dict(width=1.5)
)
centre_circle = plt.Circle((0, 0), 0.5, fc='white')
axes[0].add_artist(centre_circle)
axes[0].set_title("Sales per Category (Donut Chart)")
axes[0].axis('equal')

# Tambahkan teks total di tengah donut
axes[0].text(0, 0, f"Total\n${total_sales:,.0f}", ha='center', va='center', fontsize=20, fontweight='bold')

# --- Barplot pada axes[1] ---
sns.barplot(
    data=top_5_product,
    x='Label',
    y='Sales',
    palette=[color_dict[cat] for cat in top_5_product['Category']],
    ax=axes[1]
)
axes[1].set_title("Top 5 Products by Sales")
axes[1].set_xlabel("Product ID (with Category Prefix)")
axes[1].set_ylabel("Sales")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data region Central
store_central = store.query("Region == 'Central'")

# Filter data sub-kategori dengan profit negatif (rugi)
store_central_profit_minus = store_central[store_central['Profit'] < 0]

# Hitung total sales per sub-kategori untuk yang rugi
subcat_sales = (
    store_central_profit_minus
    .groupby('Sub-Category')['Sales']
    .sum()
    .reset_index()
    .sort_values('Sales', ascending=False)
)

# Buat figure dan axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

# === Plot 1: Scatterplot Sales vs Profit ===
sns.scatterplot(
    data=store_central,
    x='Sales',
    y='Profit',
    hue='Category',
    palette='Set2',
    ax=axes[0],
    s=100,
    edgecolor='black',
    alpha=0.7
)
axes[0].set_title('Sales vs Profit (Region: Central)', fontsize=16)
axes[0].set_xlabel('Sales ($)', fontsize=12)
axes[0].set_ylabel('Profit ($)', fontsize=12)
axes[0].grid(True)
axes[0].legend(title='Category')

# === Plot 2: Barplot Sub-Category (Profit Negatif) ===
sns.barplot(
    data=subcat_sales,
    x='Sub-Category',
    y='Sales',
    palette='Set2',
    ax=axes[1]
)
axes[1].set_title('Total Sales of Sub-Categories with Losses (Region: Central)', fontsize=16)
axes[1].set_xlabel('Sub-Category', fontsize=12)
axes[1].set_ylabel('Total Sales ($)', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y')

# Tambahkan label angka di atas bar secara akurat
for bar in axes[1].patches:
    height = bar.get_height()
    x = bar.get_x() + bar.get_width() / 2
    axes[1].text(
        x, height + 50,  # posisikan sedikit di atas bar
        f"${height:,.0f}",
        ha='center', va='bottom', fontsize=10)
    


plt.tight_layout()
plt.show()
```

***Insight***
- chairs dan table penjualannya tinggi tapi tetap merugi -> mungkin dijual terlalu murah / diskon terlalu besar / biaya produksi tinggi
- rekomendasi: lakukan audit, pertimbangkan strategi promosi

# **No.8 (Eksplorasi Data Bebas 2)**

- Dari data Superstore, buat 1 figur dengan 2 grafik dan tulis insight yang anda peroleh dari grafik tersebut

```python
repeat_customers = store.groupby("Customer ID")["Order ID"].nunique().reset_index()
repeat_customers = repeat_customers[repeat_customers["Order ID"] > 6]

repeat_customers_data = store[store['Customer ID'].isin(repeat_customers['Customer ID'])]
repeat_customers_data
```

```python
#produk yang paling sering dibeli
top_products = repeat_customers_data['Product Name'].value_counts().head(10)
top_products
```

```python
# hitung region
region_counts = repeat_customers_data.groupby("Region")["Customer ID"].nunique().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atur style
sns.set(style="whitegrid")

# Buat figure dan axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Plot kiri: Pie chart Repeat Customer per Region
axes[0].pie(
    region_counts.values, 
    labels=region_counts.index, 
    autopct='%1.1f%%', 
    colors=sns.color_palette("Set2", len(region_counts)),
    startangle=140,
    wedgeprops={'edgecolor': 'white'}
)
axes[0].set_title("Distribusi Repeat Customer (>6x Order)\nper Region", fontsize=14)

# Plot kanan: Produk Teratas yang Sering Dibeli Repeat Customer
top_products.plot(
    kind='barh', 
    color='dodgerblue', 
    ax=axes[1]
)
axes[1].set_title("Top 10 Produk Paling Sering Dibeli\nRepeat Customer (>6 Order)", fontsize=14)
axes[1].set_xlabel("Jumlah Pembelian")
axes[1].invert_yaxis()  # Supaya produk teratas ada di atas
axes[1].set_ylabel("")

# Tata letak
plt.tight_layout()
plt.show()
```

***Insight***
- dikeempat region lebih dari 20% customer melakukan repeat order
- item yang paling sering di repeat adalah item-item berhubungan dengan staple
- rekomendasi: maintain item-item staple tersebut dan boost untuk barang sisanya, karena sangat berpotensi
